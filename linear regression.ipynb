```python
# Dummy Vars & Encoders
from numpy import array
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
dataset2 = ['Pizza','Burger','Bread','Bread','Bread','Burger','Pizza','Burger']
values = array(dataset2)
print(values)
```

```python
label_encoder = LabelEncoder()              # encode the dataset2      ( textual data to numerical data
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded)
```

```python
onehot = OneHotEncoder(sparse=False)    #convert categorial data to numerical data(above values are arrange in matrix form)
integer_encoded = integer_encoded.reshape(len(integer_encoded),1)
onehot_encoded = onehot.fit_transform(integer_encoded)
print(onehot_encoded)
```

```python
#inverted_result = label_encoder.inverse_transform([argmax(onehot_encoded[0,:])])
#print(inverted_result)
# Encoding categorical data
#from sklearn.prrocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()                     # print x values
X
```

```python
#inverted_result = label_encoder.inverse_transform([argmax(onehot_encoded[0,:])])
#print(inverted_result)
 #Encoding categorical data
#from sklearn.prrocessing import LabelEncoder, OneHotEncoder
```

```python
X[:, 3] = labelencoder.fit_transform(X[:, 3])   # converting  state values to  numeric values
X
```

```python
onehotencoder = OneHotEncoder(categorical_features = [3])     #dummy values are created new coloumns are created
X = onehotencoder.fit_transform(X).toarray()
X
```

```python
# Avoiding the Dummy Variable Trap
X = X[:, 1:]                               #delete first  coloumn from above x r&d
X
```

```python
test_size = 0.33
from sklearn.model_selection import train_test_split
#pip install -U scikit-learn
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=test_size)
print('Partitioning Done!')
```

```python
# Fitting Multiple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regr.predict(X_test)
y_pred
```
